```python
import pandas as pd
```

```python
df = pd.read_csv("./ReasoningBase.csv")
df.head()
```

Inside

```python
df['Affordance:Empty'] = False
df['Affordance:Fill'] = False

df.loc[(df['Inside'] == 'Full'), 'Affordance:Empty'] = True
df.loc[(df['Inside'] == 'Empty'), 'Affordance:Fill'] = True
```

State

```python
df['Affordance:Close'] = False
df['Affordance:Open'] = False

df.loc[(df['State'] == 'Open'), 'Affordance:Close'] = True
df.loc[(df['State'] == 'Closed'), 'Affordance:Open'] = True
```

Finish

```python
df['Affordance:Unscrew'] = False
df['Affordance:Lift'] = False
df['Affordance:Swing'] = False

df.loc[(df['Finish'] == 'Snap') & (df['State'] == 'Closed'), 'Affordance:Lift'] = True
df.loc[(df['Finish'] == 'Twist') & (df['State'] == 'Closed'), 'Affordance:Unscrew'] = True
df.loc[(df['Finish'] == 'Swing-top') & (df['State'] == 'Closed'), 'Affordance:Swing'] = True
```

Metal Clip, Rubber Ring, Handle

```python
df['Affordance:PullRubber'] = False
df['Affordance:RemoveClip'] = False
df['Affordance:GrabHandle'] = False

df.loc[(df['Metal Clip']), 'Affordance:RemoveClip'] = True
df.loc[(df['Rubber Ring']), 'Affordance:PullRubber'] = True
df.loc[(df['Handle']), 'Affordance:GrabHandle'] = True
```

Closure

```python
df['Closure'].unique()
```

```python
# df['Affordance:Unscrew'] = False
# df['Affordance:Lift'] = False
df['Affordance:FlipOff'] = False
df['Affordance:FlipLid'] = False
df['Affordance:Squeeze'] = False
df['Affordance:PressDown'] = False
df['Affordance:PullUp'] = False
df['Affordance:Corkscrew'] = False
df['Affordance:RemoveSeal'] = False
# df['Affordance:Swing'] = False
#df['Affordance:GrabHandle'] = False

df.loc[(df['Closure'] == 'Snap') & (df['State'] == 'Closed'), 'Affordance:Lift'] = True
df.loc[(df['Closure'] == 'Screw') & (df['State'] == 'Closed'), 'Affordance:Unscrew'] = True
df.loc[(df['Closure'] == 'Crown') & (df['State'] == 'Closed'), 'Affordance:FlipOff'] = True
df.loc[(df['Closure'] == 'Flip-top') & (df['State'] == 'Closed'), 'Affordance:FlipLid'] = True
df.loc[(df['Closure'] == 'Trigger') & (df['State'] == 'Closed'), 'Affordance:Squeeze'] = True
df.loc[(df['Closure'] == 'Spray') & (df['State'] == 'Closed'), 'Affordance:PressDown'] = True
df.loc[(df['Closure'] == 'Push-pull') & (df['State'] == 'Closed'), 'Affordance:PullUp'] = True
df.loc[(df['Closure'] == 'Dropper') & (df['State'] == 'Closed'), 'Affordance:Squeeze'] = True
df.loc[(df['Closure'] == 'Cork') & (df['State'] == 'Closed'), 'Affordance:Corkscrew'] = True
df.loc[(df['Closure'] == 'Ropp') & (df['State'] == 'Closed'), 'Affordance:RemoveSeal'] = True
df.loc[(df['Closure'] == 'Swing-top') & (df['State'] == 'Closed'), 'Affordance:Swing'] = True
df.loc[(df['Closure'] == 'Handle') & (df['State'] == 'Closed'), 'Affordance:GrabHandle'] = True
```
